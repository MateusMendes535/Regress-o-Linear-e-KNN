```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('/content/ConsumoCo2.csv')
df.head()
```

```python
x = df[['MODELYEAR', 'ENGINESIZE']]
y = df['CO2EMISSIONS']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.3,
                                                    random_state = 45)
```

```python
#Para a Regressão Linear#

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
previssoes = linear_model.predict(x_test)

r2_score = r2_score(y_test, previssoes)
mse = mean_squared_error(y_test, previssoes)
mae = mean_absolute_error(y_test, previssoes)

metricas = {
    'r2_score': [r2_score],
    'mse': [mse],
    'mae': [mae]
}

df_metricas = pd.DataFrame(metricas)
df_metricas.rename()
```

```python
#Para o KNN#

linear_model = KNeighborsRegressor()
linear_model.fit(x_train, y_train)
previssoes = linear_model.predict(x_test)

r2_score = r2_score(y_test, previssoes)
mse = mean_squared_error(y_test, previssoes)
mae = mean_absolute_error(y_test, previssoes)

metricas = {
    'r2_score': [r2_score],
    'mse': [mse],
    'mae': [mae]
}

df_metricas = pd.DataFrame(metricas)
df_mertricas
```
